```python
def category_ratings(busniess_clusters):
    
    negcluster = busniess_clusters[busniess_clusters['stars']<3.5].reset_index()
    poscluster = busniess_clusters[busniess_clusters['stars']>=3.5].reset_index()

    pncaty={}


    for i in range(len(poscluster['categories'])):
        foodlist = poscluster['categories'][i].strip("[]").split(',')
        wt = poscluster['review_count'][i]*poscluster['stars'][i]
        for word in foodlist:
            word = word.rstrip(',')
            pncaty[word]=pncaty.get(word,0)+wt
            #print word

    for i in range(len(negcluster['categories'])):
        foodlist = negcluster['categories'][i].strip("[]").split(',')
        wt = -(negcluster['review_count'][i]*negcluster['stars'][i])
        for word in foodlist:
            word = word.rstrip(',')
            pncaty[word]=pncaty.get(word,0)+wt
    
    return pncaty
```
